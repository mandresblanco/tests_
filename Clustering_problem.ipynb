```python
df = pd.read_csv('/content/OA 5.6.csv')
X1= df.Category1
X2= df.Category2
X3= df.Category3
X4= df.Category4
X5= df.Category5
X6= df.Category6
X7= df.Category7
X8= df.Category8
X9= df.Category9
X10= df.Category10
X = []
for i in range(1,980):
    Y = (X1[i]+X2[i]+X3[i]+X4[i]+X5[i]+X6[i]+X7[i]+X8[i]+X9[i]+X10[i])/10
    X.append([X1[i],Y]) 
    X.append([X1[i],Y])
    X.append([X2[i],Y])
    X.append([X3[i],Y])
    X.append([X4[i],Y])
    X.append([X5[i],Y])
    X.append([X6[i],Y])
    X.append([X7[i],Y])
    X.append([X8[i],Y])
    X.append([X9[i],Y])
    X.append([X10[i],Y])

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

#     p.array([X1 , X2, X3 , X4, X5 , X6, X7 , X8, X9 , X10])
# # X

#     p.array([X1 , X2, X3 , X4, X5 , X6, X7 , X8, X9 , X10])
# X
# X
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python

colors = ['g.', 'r.', 'c.', 'y.','b.']

for i in range(len(X)):
    # print('Coordinate:',X[i], 'Label:', labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]],
    markersize = 10)
```
